```python
import pandas as pd
import warnings
warnings.simplefilter(action='ignore', category=FutureWarning)

# Display all columns in the DataFrame
pd.set_option("display.max_columns", None)
```

```python
df = pd.read_csv('data/Campaigns_Jan_6_2022.csv')
```

```python
def wrangle(df):
    
    df1 = df.copy()
    drop_cols = ["State", "Status", "End date", "Portfolio", "Type", "Targeting",
                 "Campaign bidding strategy", "NTB orders", "% of orders NTB", 
                 "NTB sales(USD)", "% of sales NTB", "Viewable impressions", 
                 "VCPM(USD)", "CPC(USD)"]
    
    # Filtering for only "ENABLED" ad groups
    df1 = df1[df1['State'] == "ENABLED"].drop(drop_cols, axis=1)
    df1['Top-of-search IS'] = df1['Top-of-search IS'].str.replace("<", "")
    df1['Top-of-search IS'] = df1['Top-of-search IS'].str.replace("%", "")
    df1["Top-of-search IS"] = df1["Top-of-search IS"].astype(float)
    # Changing to datetime format
    df1['Start date'] = df1['Start date'].str.replace("/", "-")
    df1['Start date'] = pd.to_datetime(df['Start date'])
    df1 = df1.round(2)
    df1.sort_values("ACOS", ascending=False, inplace=True)
    
    return df1

df1 =  wrangle(df)
    
df1.head(15)
```
