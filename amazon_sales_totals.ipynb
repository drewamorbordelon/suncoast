```python
import pandas as pd
pd.set_option('display.max_columns', None)
```

```python
year = "data/2021Jan7-2022Jan6CustomUnifiedTransaction.csv"
monthly_csv = "data/2021Dec7-2022Jan6CustomUnifiedTransaction.csv"
previous_csv = "data/2021Nov7-2021Dec6CustomUnifiedTransaction.csv"




raw = pd.read_csv(year, header=0, skiprows=7, skipfooter=0)
monthly = pd.read_csv(monthly_csv, header=0, skiprows=7, skipfooter=0)
previous = pd.read_csv(previous_csv, header=0, skiprows=7, skipfooter=0)

units_sold = raw['description'].value_counts().to_frame()
month_sales = monthly['description'].value_counts().to_frame()
previous_sales = previous['description'].value_counts().to_frame() 
month_sales.columns = ['Current Month']
# previous_sales.columns['Previous Month']
units_sold.columns = ['YoY']
units_sold['Monthly_Avg'] = (units_sold['YoY'] / 12).round(0).astype(int)
units_sold['Weekly_Avg'] = (units_sold['Monthly_Avg'] / 4).astype(int)
units_sold['Daily_Avg'] = (units_sold['Weekly_Avg'] / 7).astype(int)

# month_sales = month_sales['description'].value_counts().to_frame()
comps = pd.merge(units_sold, month_sales, left_index=True, right_index=True)
comps = pd.merge(comps, previous_sales, left_index=True, right_index=True)

comps = comps.rename(columns={"Current Month": "Current_Month", 'description': 'Previous_Month'})
# comps.columns = ["YoY", 'Monthly_Avg', "Daily_Avg", "Current_Month", "Previous_Month"]
# comps = comps.head(25)

december = pd.ExcelWriter('excel_output/amazonSales_december2021.xlsx')

# Write each dataframe to a different worksheet.
comps.to_excel(december, sheet_name='Total Sales')
# amz_data_totals.to_excel(december, sheet_name='Totals', index=False)
comps.head(60)
# print(comps.shape)
```
