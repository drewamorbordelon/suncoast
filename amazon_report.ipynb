```python

import pandas as pd
pd.set_option('display.max_columns', None)

csv = "data/Amazon_2021DecMonthlyUnifiedTransaction.csv"

def amazon_report(csv: str):
    df = pd.read_csv(csv, header=None, skiprows=7, skipfooter=0)

    # Make row1 = column titles
    df.columns = df.iloc[0]

    # Delete row1 since previous code duplicated row1 and made the column titles
    df = df[1:]

    # What we want the columns to be in the new dataframe
    col_names = ['date/time', 'settlement id', 'type', 'order id', 'sku', 'description',
                 'quantity', 'marketplace', 'account type', 'fulfillment', 'order city',
                 'order state', 'order postal', 'tax collection model', 'product sales',
                 'product sales tax', 'shipping credits', 'shipping credits tax',
                 'gift wrap credits', 'giftwrap credits tax', 'promotional rebates',
                 'promotional rebates tax', 'marketplace withheld tax', 'selling fees',
                 'fba fees', 'other transaction fees', 'other', 'total']

    df.columns = ['date/time', 'settlement id', 'type', 'order id','sku', 'description', 
                  'quantity', 'marketplace', 'account type', 'fulfillment', 'order city',
                  'order state', 'order postal', 'tax collection model', 'product sales', 
                  'product sales tax', 'shipping credits', 'shipping credits tax', 
                  'gift wrap credits', 'giftwrap credits tax', 'promotional rebates', 
                  'promotional rebates tax', 'marketplace withheld tax', 'selling fees', 
                  'fba fees', 'other transaction fees', 'other', 'total']
                  ############### Drop these for now
    #               'nan', 'Amazon', 'Reimb', 'Fee', 'Payment', 'to SSG', 'Adj', 'Storage', 'Removal', 'nan', 'Adjustment', 'nan', 'nan']

    df_updated = df[col_names].copy()
    df_updated['Date'] = pd.to_datetime(df_updated['date/time'])
    df_updated.head()
    # Exchange values between 2 columns
    df_updated['date/time'] = df_updated['Date']

    # # Drop date column
    df_updated.drop(['Date'], axis=1, inplace=True)
    col_names_new = ['date', 'settlement id', 'type', 'order id', 'sku', 'description',
                 'quantity', 'marketplace', 'account type', 'fulfillment', 'order city',
                 'order state', 'order postal', 'tax collection model', 'product sales',
                 'product sales tax', 'shipping credits', 'shipping credits tax',
                 'gift wrap credits', 'giftwrap credits tax', 'promotional rebates',
                 'promotional rebates tax', 'marketplace withheld tax', 'selling fees',
                 'fba fees', 'other transaction fees', 'other', 'total']
    df_updated.columns = col_names_new
    
    df['date/time'] = pd.to_datetime(df['date/time'])
    df_copy = df[["date/time", 'product sales',
       'product sales tax', 'shipping credits', 'shipping credits tax',
       'gift wrap credits', 'giftwrap credits tax', 'promotional rebates',
       'promotional rebates tax', 'marketplace withheld tax', 'selling fees',
       'fba fees']].copy()
    df_copy.columns = ["date", 'product sales',
       'product sales tax', 'shipping credits', 'shipping credits tax',
       'gift wrap credits', 'giftwrap credits tax', 'promotional rebates',
       'promotional rebates tax', 'marketplace withheld tax', 'selling fees',
       'fba fees']
    
    df_updated['other'] = df_updated['other'].str.replace(',', '')
    df_updated['other'] = df_updated['other'].str.replace(')', '')
    df_updated['other'] = df_updated['other'].str.replace('(', '')
    df_updated['other'] = df_updated['other'].astype(float)

    df_updated['other transaction fees'] = df_updated['other transaction fees'].str.replace(',', '')
    df_updated['other transaction fees'] = df_updated['other transaction fees'].str.replace(')', '')
    df_updated['other transaction fees'] = df_updated['other transaction fees'].str.replace('(', '')
    df_updated['other transaction fees'] = df_updated['other transaction fees'].astype(float)

    df_updated['total'] = df_updated['total'].str.replace(',', '')
    df_updated['total'] = df_updated['total'].str.replace(')', '')
    df_updated['total'] = df_updated['total'].str.replace('(', '')
    df_updated['total'] = df_updated['total'].astype(float)

    df_updated['product sales'] = df_updated['product sales'].str.replace(',', '')
    df_updated['product sales'] = df_updated['product sales'].str.replace(')', '')
    df_updated['product sales'] = df_updated['product sales'].str.replace('(', '')
    df_updated['product sales'] = df_updated['product sales'].astype(float)

    df_updated['product sales tax'] = df_updated['product sales tax'].str.replace(',', '')
    df_updated['product sales tax'] = df_updated['product sales tax'].str.replace(')', '')
    df_updated['product sales tax'] = df_updated['product sales tax'].str.replace('(', '')
    df_updated['product sales tax'] = df_updated['product sales tax'].astype(float)

    df_updated['shipping credits'] = df_updated['shipping credits'].str.replace(',', '')
    df_updated['shipping credits'] = df_updated['shipping credits'].str.replace(')', '')
    df_updated['shipping credits'] = df_updated['shipping credits'].str.replace('(', '')
    df_updated['shipping credits'] = df_updated['shipping credits'].astype(float)

    df_updated['shipping credits tax'] = df_updated['shipping credits tax'].str.replace(',', '')
    df_updated['shipping credits tax'] = df_updated['shipping credits tax'].str.replace(')', '')
    df_updated['shipping credits tax'] = df_updated['shipping credits tax'].str.replace('(', '')
    df_updated['shipping credits tax'] = df_updated['shipping credits tax'].astype(float)

    df_updated['gift wrap credits'] = df_updated['gift wrap credits'].str.replace(',', '')
    df_updated['gift wrap credits'] = df_updated['gift wrap credits'].str.replace(')', '')
    df_updated['gift wrap credits'] = df_updated['gift wrap credits'].str.replace('(', '')
    df_updated['gift wrap credits'] = df_updated['gift wrap credits'].astype(float)

    df_updated['giftwrap credits tax'] = df_updated['giftwrap credits tax'].str.replace(',', '')
    df_updated['giftwrap credits tax'] = df_updated['giftwrap credits tax'].str.replace(')', '')
    df_updated['giftwrap credits tax'] = df_updated['giftwrap credits tax'].str.replace('(', '')
    df_updated['giftwrap credits tax'] = df_updated['giftwrap credits tax'].astype(float)

    df_updated['promotional rebates'] = df_updated['promotional rebates'].str.replace(',', '')
    df_updated['promotional rebates'] = df_updated['promotional rebates'].str.replace(')', '')
    df_updated['promotional rebates'] = df_updated['promotional rebates'].str.replace('(', '')
    df_updated['promotional rebates'] = df_updated['promotional rebates'].astype(float)

    df_updated['promotional rebates tax'] = df_updated['promotional rebates tax'].str.replace(',', '')
    df_updated['promotional rebates tax'] = df_updated['promotional rebates tax'].str.replace(')', '')
    df_updated['promotional rebates tax'] = df_updated['promotional rebates tax'].str.replace('(', '')
    df_updated['promotional rebates tax'] = df_updated['promotional rebates tax'].str.replace('-', '')
    df_updated['promotional rebates tax'] = df_updated['promotional rebates tax'].astype(float)

    df_updated['marketplace withheld tax'] = df_updated['marketplace withheld tax'].str.replace(',', '')
    df_updated['marketplace withheld tax'] = df_updated['marketplace withheld tax'].str.replace(')', '')
    df_updated['marketplace withheld tax'] = df_updated['marketplace withheld tax'].str.replace('(', '')
    df_updated['marketplace withheld tax'] = df_updated['marketplace withheld tax'].astype(float)

    df_updated['selling fees'] = df_updated['selling fees'].str.replace(',', '')
    df_updated['selling fees'] = df_updated['selling fees'].str.replace(')', '')
    df_updated['selling fees'] = df_updated['selling fees'].str.replace('(', '')
    df_updated['selling fees'] = df_updated['selling fees'].astype(float)

    df_updated['fba fees'] = df_updated['fba fees'].str.replace(',', '')
    df_updated['fba fees'] = df_updated['fba fees'].str.replace(')', '')
    df_updated['fba fees'] = df_updated['fba fees'].str.replace('(', '')
    df_updated['fba fees'] = df_updated['fba fees'].astype(float)

    Ship_to_Amazon = df_updated[df_updated['description'].str.contains('FBA Amazon-Partnered Carrier Shipment Fee|ReturnPostageBilling', na=False)]
    Ship_to_Amazon['Ship to Amazon Fee'] = Ship_to_Amazon['other'].copy()
    FBA_inventory_reimb = df_updated[df_updated['type'].str.contains('Adjustment|Refund', na=False)]
    FBA_inventory_reimb['FBA Inventory Reimb'] = FBA_inventory_reimb['other'].copy()
    Monthly_Fee = df_updated[df_updated['description'].str.contains('Subscription', na=False)]
    Monthly_Fee['Monthly_Amz_Fee'] = Monthly_Fee['other'].copy()
    Transfer_to_SSG = df_updated[df_updated['description'].str.contains('To your account ending in: 603', na=False)]
    Transfer_to_SSG['Transfer_to_SSG_603_Acct'] = Transfer_to_SSG['other'].copy()
    FBA_invStorageFee = df_updated[df_updated['description'].str.contains('FBA Inventory Storage Fee', na=False)]
    FBA_invStorageFee['FBA Inventory Storage Fee'] = FBA_invStorageFee['other'].copy()
    FBA_removalOrder = df_updated[df_updated['description'].str.contains('FBA Removal Order: Return Fee', na=False)]
    FBA_removalOrder['FBA_RemovalOrder_ReturnFee'] = FBA_removalOrder['other'].copy()
    Advertising = df_updated[df_updated['description'].str.contains('Cost of Advertising', na=False)]
    Advertising['Cost of Advertising'] = Advertising['other transaction fees'].copy()

    ls_subitems = [Ship_to_Amazon, FBA_inventory_reimb, Transfer_to_SSG, Monthly_Fee, FBA_invStorageFee, FBA_removalOrder, Advertising]

    data = pd.concat(ls_subitems)
    amz_data = data[['date', 'settlement id', 'type', 'order id', 'description', 'other transaction fees','other', 'total', 
                     'product sales','product sales tax', 'shipping credits', 'shipping credits tax','gift wrap credits', 
                     'giftwrap credits tax', 'promotional rebates','promotional rebates tax', 'marketplace withheld tax', 
                     'selling fees','fba fees',
                     'Ship to Amazon Fee', 'FBA Inventory Reimb', 'Monthly_Amz_Fee', 'Transfer_to_SSG_603_Acct', 'FBA Inventory Storage Fee', 
                     'FBA_RemovalOrder_ReturnFee', 'Cost of Advertising']].copy()


    amz_data = amz_data.fillna(0)
    amz_data = amz_data.sort_index()
    amz_data.drop(['date', 'product sales', 'product sales tax', 'shipping credits',
           'shipping credits tax', 'gift wrap credits', 'giftwrap credits tax',
           'promotional rebates', 'promotional rebates tax',
           'marketplace withheld tax', 'selling fees', 'fba fees'], axis=1, inplace=True)

    amz_data = df_copy.join(amz_data)


    amz_data['product sales'] = amz_data['product sales'].str.replace(',', '')
    amz_data['product sales'] = amz_data['product sales'].str.replace(')', '')
    amz_data['product sales'] = amz_data['product sales'].str.replace('(', '')
    amz_data['product sales'] = amz_data['product sales'].astype(float)

    amz_data['product sales tax'] = amz_data['product sales tax'].str.replace(',', '')
    amz_data['product sales tax'] = amz_data['product sales tax'].str.replace(')', '')
    amz_data['product sales tax'] = amz_data['product sales tax'].str.replace('(', '')
    amz_data['product sales tax'] = amz_data['product sales tax'].astype(float)

    amz_data['shipping credits'] = amz_data['shipping credits'].str.replace(',', '')
    amz_data['shipping credits'] = amz_data['shipping credits'].str.replace(')', '')
    amz_data['shipping credits'] = amz_data['shipping credits'].str.replace('(', '')
    amz_data['shipping credits'] = amz_data['shipping credits'].astype(float)

    amz_data['shipping credits tax'] = amz_data['shipping credits tax'].str.replace(',', '')
    amz_data['shipping credits tax'] = amz_data['shipping credits tax'].str.replace(')', '')
    amz_data['shipping credits tax'] = amz_data['shipping credits tax'].str.replace('(', '')
    amz_data['shipping credits tax'] = amz_data['shipping credits tax'].astype(float)

    amz_data['gift wrap credits'] = amz_data['gift wrap credits'].str.replace(',', '')
    amz_data['gift wrap credits'] = amz_data['gift wrap credits'].str.replace(')', '')
    amz_data['gift wrap credits'] = amz_data['gift wrap credits'].str.replace('(', '')
    amz_data['gift wrap credits'] = amz_data['gift wrap credits'].astype(float)

    amz_data['giftwrap credits tax'] = amz_data['giftwrap credits tax'].str.replace(',', '')
    amz_data['giftwrap credits tax'] = amz_data['giftwrap credits tax'].str.replace(')', '')
    amz_data['giftwrap credits tax'] = amz_data['giftwrap credits tax'].str.replace('(', '')
    amz_data['giftwrap credits tax'] = amz_data['giftwrap credits tax'].astype(float)

    amz_data['promotional rebates'] = amz_data['promotional rebates'].str.replace(',', '')
    amz_data['promotional rebates'] = amz_data['promotional rebates'].str.replace(')', '')
    amz_data['promotional rebates'] = amz_data['promotional rebates'].str.replace('(', '')
    amz_data['promotional rebates'] = amz_data['promotional rebates'].astype(float)

    amz_data['promotional rebates tax'] = amz_data['promotional rebates tax'].str.replace(',', '')
    amz_data['promotional rebates tax'] = amz_data['promotional rebates tax'].str.replace(')', '')
    amz_data['promotional rebates tax'] = amz_data['promotional rebates tax'].str.replace('(', '')
    amz_data['promotional rebates tax'] = amz_data['promotional rebates tax'].astype(float)

    amz_data['marketplace withheld tax'] = amz_data['marketplace withheld tax'].str.replace(',', '')
    amz_data['marketplace withheld tax'] = amz_data['marketplace withheld tax'].str.replace(')', '')
    amz_data['marketplace withheld tax'] = amz_data['marketplace withheld tax'].str.replace('(', '')
    amz_data['marketplace withheld tax'] = amz_data['marketplace withheld tax'].astype(float)

    amz_data['selling fees'] = amz_data['selling fees'].str.replace(',', '')
    amz_data['selling fees'] = amz_data['selling fees'].str.replace(')', '')
    amz_data['selling fees'] = amz_data['selling fees'].str.replace('(', '')
    amz_data['selling fees'] = amz_data['selling fees'].astype(float)

    amz_data['fba fees'] = amz_data['fba fees'].str.replace(',', '')
    amz_data['fba fees'] = amz_data['fba fees'].str.replace(')', '')
    amz_data['fba fees'] = amz_data['fba fees'].str.replace('(', '')
    amz_data['fba fees'] = amz_data['fba fees'].astype(float)
    
    amz_data = amz_data.fillna(0)

    amz_data['ProductSales_Cum_Total'] = amz_data['product sales'].cumsum(skipna=True)
    amz_data['ProductSalesTax_Cum_Total'] = amz_data['product sales tax'].cumsum(skipna=True)
    amz_data['ShippingCredits_Cum_Total'] = amz_data['shipping credits'].cumsum(skipna=True)
    amz_data['ShippingCredits_Tax_Cum_Total'] = amz_data['shipping credits tax'].cumsum(skipna=True)
    amz_data['GiftWrapCredits_Cum_Total'] = amz_data['gift wrap credits'].cumsum(skipna=True)
    amz_data['GiftWrapCredits_Tax_Cum_Total'] = amz_data['giftwrap credits tax'].cumsum(skipna=True)
    amz_data['PromoRebates_Cum_Total'] = amz_data['promotional rebates'].cumsum(skipna=True)
    amz_data['PromoRebates_Tax_Cum_Total'] = amz_data['promotional rebates tax'].cumsum(skipna=True)
    amz_data['MarketPlaceWithheld_Tax_Cum_Total'] = amz_data['marketplace withheld tax'].cumsum(skipna=True)
    amz_data['SellingFees_Cum_Total'] = amz_data['selling fees'].cumsum(skipna=True)
    amz_data['FBAFees_Cum_Total'] = amz_data['fba fees'].cumsum(skipna=True)


    amz_data['Ship_Cum_Total'] = amz_data['Ship to Amazon Fee'].cumsum(skipna=True)
    amz_data['FBA_InvenReimb_Cum_Total'] = amz_data['FBA Inventory Reimb'].cumsum(skipna=True)
    amz_data['AmazonFee_Cum_Total'] = amz_data['Monthly_Amz_Fee'].cumsum(skipna=True)
    amz_data['TransferSSG_Cum_Total'] = amz_data['Transfer_to_SSG_603_Acct'].cumsum(skipna=True)
    amz_data['FBA_StorageCum_Total'] = amz_data['FBA Inventory Storage Fee'].cumsum(skipna=True)

    # amz_data['FBA_RemovalOrder_ReturnFee_Cum_Total'] = amz_data['FBA_RemovalOrder_ReturnFee'].cumsum(skipna=True)

    amz_data['FBA_RemovalOrder_ReturnFee_Cum_Total'] = amz_data['FBA_RemovalOrder_ReturnFee'].cumsum(skipna=True)
    amz_data['AdvertCostCum_Total'] = amz_data['Cost of Advertising'].cumsum(skipna=True)

    # Comparing totals together
    amz_data['OtherTotal'] = amz_data['other transaction fees'] + amz_data['other']
    amz_data['Other_FeeTotals'] = amz_data['OtherTotal'].cumsum(skipna=True)

    # Making a column to compare to
    amz_data['CompareTotal'] = amz_data['Ship_Cum_Total']+amz_data['FBA_InvenReimb_Cum_Total']+amz_data['AmazonFee_Cum_Total']+amz_data['TransferSSG_Cum_Total']+amz_data['FBA_StorageCum_Total']+amz_data['FBA_RemovalOrder_ReturnFee_Cum_Total']+amz_data['AdvertCostCum_Total']
    
#     amz_data['description'].value_counts()
    amz_df_totals = amz_data[['date', 'ProductSales_Cum_Total',
                           'ProductSalesTax_Cum_Total', 'ShippingCredits_Cum_Total',
                           'ShippingCredits_Tax_Cum_Total', 'GiftWrapCredits_Cum_Total',
                           'GiftWrapCredits_Tax_Cum_Total', 'PromoRebates_Cum_Total',
                           'PromoRebates_Tax_Cum_Total', 'MarketPlaceWithheld_Tax_Cum_Total',
                           'SellingFees_Cum_Total', 'FBAFees_Cum_Total', 'Ship_Cum_Total',
                           'FBA_InvenReimb_Cum_Total', 'AmazonFee_Cum_Total',
                           'TransferSSG_Cum_Total', 'FBA_StorageCum_Total',
                           'FBA_RemovalOrder_ReturnFee_Cum_Total', 'AdvertCostCum_Total', 'Other_FeeTotals', 'CompareTotal']].copy()

    amz_df_totals.columns = ['date', 'ProductSales',
                               'ProductSalesTax', 'ShippingCredits',
                               'ShippingCredits_Tax', 'GiftWrapCredits',
                               'GiftWrapCredits_Tax', 'PromoRebates',
                               'PromoRebates_Tax', 'MarketPlaceWithheld_Tax',
                               'SellingFees', 'FBAFees', 'Shipping',
                               'FBA_InventoryReimbursement', 'AmazonFee',
                               'TransferSSG_603', 'FBA_StorageFee',
                               'FBA_RemovalOrder_ReturnFee', 'AdvertisingCost', 'Other_FeeTotals', 'CompareTotal']  #  , 'OtherTotals', 'Totals'
    amz_data_totals = amz_df_totals.tail(1)

    amz_data_totals = amz_data_totals.set_index('date')
    amz_data_totals = amz_data_totals.rename_axis('index')
    amz_data_totals = amz_data_totals.T
    amz_data_totals.reset_index(inplace=True)
    amz_data_totals.rename(columns={'index':'amazon'}, inplace=True)

    amz_data_totals = amz_data_totals.round(2)
    
    december = pd.ExcelWriter('excel_output/amazon_december2021.xlsx')

    # Write each dataframe to a different worksheet.
    amz_data.to_excel(december, sheet_name='Sheet1')
    amz_data_totals.to_excel(december, sheet_name='Totals', index=False)


    # Close the Pandas Excel writer and output the Excel file.
    december.save()
    
    return amz_data_totals
```
