```python
amz_data = amz_data.fillna(0)

amz_data['ProductSales_Cum_Total'] = amz_data['product sales'].cumsum(skipna=True)
amz_data['ProductSalesTax_Cum_Total'] = amz_data['product sales tax'].cumsum(skipna=True)
amz_data['ShippingCredits_Cum_Total'] = amz_data['shipping credits'].cumsum(skipna=True)
amz_data['ShippingCredits_Tax_Cum_Total'] = amz_data['shipping credits tax'].cumsum(skipna=True)
amz_data['GiftWrapCredits_Cum_Total'] = amz_data['gift wrap credits'].cumsum(skipna=True)
amz_data['GiftWrapCredits_Tax_Cum_Total'] = amz_data['giftwrap credits tax'].cumsum(skipna=True)
amz_data['PromoRebates_Cum_Total'] = amz_data['promotional rebates'].cumsum(skipna=True)
amz_data['PromoRebates_Tax_Cum_Total'] = amz_data['promotional rebates tax'].cumsum(skipna=True)
amz_data['MarketPlaceWithheld_Tax_Cum_Total'] = amz_data['marketplace withheld tax'].cumsum(skipna=True)
amz_data['SellingFees_Cum_Total'] = amz_data['selling fees'].cumsum(skipna=True)
amz_data['FBAFees_Cum_Total'] = amz_data['fba fees'].cumsum(skipna=True)


amz_data['Ship_Cum_Total'] = amz_data['Ship to Amazon Fee'].cumsum(skipna=True)
amz_data['FBA_InvenReimb_Cum_Total'] = amz_data['FBA Inventory Reimb'].cumsum(skipna=True)
amz_data['AmazonFee_Cum_Total'] = amz_data['Monthly_Amz_Fee'].cumsum(skipna=True)
amz_data['TransferSSG_Cum_Total'] = amz_data['Transfer_to_SSG_603_Acct'].cumsum(skipna=True)
amz_data['FBA_StorageCum_Total'] = amz_data['FBA Inventory Storage Fee'].cumsum(skipna=True)

# amz_data['FBA_RemovalOrder_ReturnFee_Cum_Total'] = amz_data['FBA_RemovalOrder_ReturnFee'].cumsum(skipna=True)

amz_data['FBA_RemovalOrder_ReturnFee_Cum_Total'] = amz_data['FBA_RemovalOrder_ReturnFee'].cumsum(skipna=True)
amz_data['AdvertCostCum_Total'] = amz_data['Cost of Advertising'].cumsum(skipna=True)

# Comparing totals together
amz_data['OtherTotal'] = amz_data['other transaction fees'] + amz_data['other']
amz_data['Other_FeeTotals'] = amz_data['OtherTotal'].cumsum(skipna=True)

# Making a column to compare to
amz_data['CompareTotal'] = amz_data['Ship_Cum_Total']+amz_data['FBA_InvenReimb_Cum_Total']+amz_data['AmazonFee_Cum_Total']+amz_data['TransferSSG_Cum_Total']+amz_data['FBA_StorageCum_Total']+amz_data['FBA_RemovalOrder_ReturnFee_Cum_Total']+amz_data['AdvertCostCum_Total']
amz_data.tail()
```
