```python
import pandas as pd
pd.set_option('display.max_columns', None)

#path
# june_df = "data/etsy_statement_2021_6.csv"
# july_df = "data/etsy_statement_2021_7(1).csv"
# august_df = "data/etsy_statement_2021_8.csv"
# sept_df = "data/etsy_statement_2021_9.csv"
# oct_df = "data/etsy_statement_2021_10.csv"
# nov_df = "data/etsy_statement_2021_11.csv"
dec_df = "data/etsy_statement_2021_12.csv"


# add "data" to beginning of the path

df = pd.read_csv(dec_df, header=None, skiprows=0, skipfooter=0)

# df = pd.read_csv(july_df, header=None, skiprows=0, skipfooter=0)
# df = pd.read_csv(august_df, header=None, skiprows=0, skipfooter=0)
# df = pd.read_csv(sept_df, header=None, skiprows=0, skipfooter=0)
# df = pd.read_csv(oct_df, header=None, skiprows=0, skipfooter=0)
df1 = df.copy()

# Make row1 = column titles
df.columns = df.iloc[0]
df = df[1:]
df = df.iloc[::-1]
df.reset_index(inplace=True)
df.drop(['index'], inplace=True, axis=1)

df['Date'] = pd.to_datetime(df['Date'])
df.rename({"Date" : "date"}, inplace=True, axis=1)

df['Amount'] = df['Amount'].str.replace('--', '0')
df['Amount'] = df['Amount'].str.replace(')', '')
df['Amount'] = df['Amount'].str.replace('(', '')
df['Amount'] = df['Amount'].str.replace('$', '')
df['Amount'] = df['Amount'].astype(float)
# df['Amount'] = df['Amount'] * -1

df['Fees & Taxes'] = df['Fees & Taxes'].str.replace('--', '0')
df['Fees & Taxes'] = df['Fees & Taxes'].str.replace(')', '')
df['Fees & Taxes'] = df['Fees & Taxes'].str.replace('(', '')
df['Fees & Taxes'] = df['Fees & Taxes'].str.replace('$', '')
df['Fees & Taxes'] = df['Fees & Taxes'].astype(float)
df['Fees & Taxes'] = df['Fees & Taxes'] 

df.drop(['Currency'], axis=1, inplace=True)
# df['Net'] = df['Net'].str.strip("($)")


fee = df[df['Type'].str.contains('Listing', na=False)]
df['listing_fee'] = fee['Fees & Taxes'].copy()
df['listing_fee'].fillna(0, inplace=True)
df['listing_fee'] = df['listing_fee'].astype(float)

tax = df[df['Type'].str.contains('Sale', na=False)]
df['tax'] = tax['Fees & Taxes'].copy()
df['tax'].fillna(0, inplace=True)

deposit = df[df['Type'].str.contains('Deposit', na=False)]
df['deposit'] = deposit['Title'].copy()
df['deposit'] = df['deposit'].str.strip("($)")
df['deposit'] = df['deposit'].str.extract('([0-9]+[,./]*[0-9]*)')
df['deposit'] = df['deposit'].astype(float)
df['deposit'].fillna(0, inplace=True)

marketing = df[df['Type'].str.contains('Marketing', na=False)]
df['marketing_fee'] = marketing['Fees & Taxes'].copy()
df['marketing_fee'].fillna(0, inplace=True)

transaction = df[df['Type'].str.contains('Transaction', na=False)]
df['transaction_fee'] = transaction['Fees & Taxes'].copy()
df['transaction_fee'].fillna(0, inplace=True)

#if the blanks are nan will need this line first
df.fillna(0, inplace=True)
df['Net'] = df['Net'].str.strip("($)")
net_sale = df[df['Type'].str.contains('Sale', na=False)]
df['net'] = net_sale['Net'].copy()
df.fillna(0, inplace=True)
df['net'] = df['net'].astype(float)
df['Net1']=df['transaction_fee']+df['listing_fee'] 
df.fillna(0, inplace=True)
df['Net2'] = df['Net1'] + df['net']
df['Net'] = df['Net2']
df.drop(['Net1', 'net', 'Net2'], axis=1, inplace=True)

df.rename({"Amount" : "Payment"}, inplace=True, axis=1)

print(df.columns)
df.head(60)
```

```python
df['Payout_to_SSG603_Etsy'] = df['deposit'].cumsum(skipna=True)
df['Listing_Fee'] = df['listing_fee'].cumsum(skipna=True)
df['Marketing_Fee'] = df['marketing_fee'].cumsum(skipna=True)

# df['Shipping_Fee'] = df['']


df['Transaction_Fee'] = df['transaction_fee'].cumsum(skipna=True)
df['Withheld_Tax_Etsy'] = df['tax'].cumsum(skipna=True)
df['GrossNet'] = df['Net'].cumsum(skipna=True)
df['Payment_flat'] = df['Payment'].cumsum(skipna=True)
df['Payment_shipping'] = df['transaction_fee'].cumsum(skipna=True)
df['PaymentNet'] = df['Payment_flat'] + df['Payment_shipping']
# payment plus the shipping
# df['GrossLessTax'] = df['PaymentNet'] + df['Withheld_Tax_Etsy'] + df['Listing_Fee'] + df['Marketing_Fee'] + df['Transaction_Fee']
```
